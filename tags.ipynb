```python
#подсчет количества тегов
import pandas as pd
filename = 'song_info.json'
data_df = pd.read_json('song_info.json', lines=True)

li = []
for i in data_df["tags"]:
    for t in i:
        li.append(t)
ints_list1 = list(set(li)) 
print(ints_list1)

counter = {}
for elem in li:
    counter[elem] = counter.get(elem, 0) + 1
doubles = {element: count for element, count in counter.items() if count > 1}
print(doubles)

sorted_tuple = sorted(doubles.items(), key=lambda x: x[1])
print(sorted_tuple)
```
